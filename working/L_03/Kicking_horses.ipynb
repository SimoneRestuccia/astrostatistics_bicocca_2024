```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from scipy.stats import poisson
#from scipy.stats import norm
from tqdm import tqdm
```

```python
def poisson_dist(mean, deaths):
    dist = poisson(mean)
    return dist.pmf(deaths)
```

```python
frequency = np.array([109, 65, 22, 3, 1])/200
deaths = np.array([0, 1, 2, 3, 4])
```

```python
weighted_mean = np.average(deaths, weights = frequency)
aritmetic_mean = np.mean(deaths)
```

```python
plt.scatter(deaths, frequency, color = 'red')
plt.plot(deaths, poisson_dist(aritmetic_mean, deaths), color='royalblue', label='Aritmetic mean')
plt.plot(deaths, poisson_dist(weighted_mean, deaths), color='black', label='Weighted mean')
plt.xlabel('Number of deaths')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
