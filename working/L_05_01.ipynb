```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats, interpolate
from scipy.interpolate import CubicSpline
from tqdm import tqdm
from scipy.stats import norm
```

```python
N = 10
mu = 1
sigma = 0.2
```

```python
s = np.random.normal(mu, sigma, N)
s
```

```python
M = 100000
for i in range(len(s)):
    x = np.random.normal(s[i], sigma, M)
    plt.hist(x, histtype='step', bins = int(np.round(M**0.5)), density = True)
plt.show()
```

```python
for i in tqdm(range(len(s))):
    z = np.linspace(0, 2, 1000)
    dist = norm(s[i], sigma)
    plt.plot(z, dist.pdf(z))
plt.show()
```

```python
def f(x, mu):
    return (1/(sigma*((2*np.pi)**0.5)))*(np.exp(-((x-mu)**0.5)/(2*(sigma**0.5))))
u = np.linspace(0, 2, 100)
fun = np.zeros(len(u))
for i in tqdm(range(len(u))):
    fun[i] = f(s[0], u[i])
plt.plot(u, fun)
plt.show()
```
