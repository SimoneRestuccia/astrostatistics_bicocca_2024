```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 10000
x = np.random.uniform(0.1, 10, N)
```

```python
plt.hist(x, density = True, histtype='step', bins = int(np.round(N**0.5)), color='black')
plt.xlabel('x')
plt.ylabel('pdf(x)')
plt.title('pdf(x)')
plt.show()
```

```python
y = np.log10(x)
```

```python
def pdf(y):
    return (np.log(10)/9.9)*(10**y)
```

```python
y_sample = np.linspace(-1, 1, N)
plt.hist(y, density=True, histtype='step', bins = int(np.round(N**0.5)), color = 'gray')
plt.plot(y_sample, pdf(y_sample), 'k', label='Theoretical pdf(y)')
plt.xlabel('y')
plt.ylabel('pdf(y)')
plt.title('pdf(y)')
plt.legend()
plt.show()
```

```python
print('The log of the mean of x is: %f;\nThe mean of y is: %f.\n\nThe log of the medeian of x is: %f;\nThe median of y is: %f.' %(np.log10(np.mean(x)), np.mean(y), np.log10(np.median(x)), np.median(y))) 
```
