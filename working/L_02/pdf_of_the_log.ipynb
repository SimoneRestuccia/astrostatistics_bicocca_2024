### Pdf of the log
1. Use `numpy` to draw $N$ (where $N$ is large...) random samples from a uniform distribution between $0.1$ and $10$, and store these samples as $x$.
2. Use matplotlib to make a histogram of these samples.
3. Compute the base-10 log of your array $x$, and store this as $y$.
4. Make another histogram for $y$. Using the equation to transform probability distributions, write what the theoretical pdf of $y$ is, and overplot it onto your histogram.
5. Compute the log of the mean of $x$ and the mean of $y$. Now compute the log of the median of $x$ and the median of $y$. 

You should note that the means are different, but the medians (as it is a cumulative statistic) are the same. The mean is affected by the scale of the sample values, but the median only depends on the ordering of the samples. Monotonic transformations (like taking the log) do not change the ordering of samples.
 

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 10000
x = np.random.uniform(0.1, 10, N)  #Sampling random x from a uniform distribution
```

```python
#Making a histogram of the x samples
plt.hist(x, density = True, histtype='step', bins = int(np.round(N**0.5)), color='black')
plt.xlabel('x')
plt.ylabel('pdf(x)')
plt.title('pdf(x)')
plt.show()
```

```python
y = np.log10(x)  
```

```python
#Defining the distribution of y using the equation to transform probability distribution
def pdf(y):
    return (np.log(10)/9.9)*(10**y)
```

```python
#Plotting the histogram of y and its theoretical distribution
y_sample = np.linspace(-1, 1, N)
plt.hist(y, density=True, histtype='step', bins = int(np.round(N**0.5)), color = 'gray')
plt.plot(y_sample, pdf(y_sample), 'k', label='Theoretical pdf(y)')
plt.xlabel('y')
plt.ylabel('pdf(y)')
plt.title('pdf(y)')
plt.legend()
plt.show()
```

```python
print('The log of the mean of x is: %f;\nThe mean of y is: %f.\n\nThe log of the medeian of x is: %f;\nThe median of y is: %f.' 
      %(np.log10(np.mean(x)), np.mean(y), np.log10(np.median(x)), np.median(y))) 
```
