```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats, interpolate
from scipy.interpolate import CubicSpline
from tqdm import tqdm
from scipy.stats import uniform
from sklearn.mixture import GaussianMixture
from astropy.visualization.hist import hist as fancyhist
from sklearn.neighbors import KernelDensity
```

- 1 = clear   
- 0 = cloudly

```python
N = 100000
p = np.zeros(N)
p[0] = 1
i = 1
a = 1
n = []
count = []
n.append(1)
count.append(1)
while i != N-1:
    r = uniform.rvs(size = 1)
    if p[i] == 1 :
        if r <=0.9 :
            p[i+1] = 1
            a = a+1
        else:
            p[i+1] = 0
    elif p[i] == 0 :
        if r <=0.5 :
            p[i+1] = 1
            a = a+1
        else:
            p[i+1] = 0
    i = i + 1
    count.append(a/i)
    n.append(i)
n.append(N)
count.append(a/N)
```

```python
sunny = p[p==1]
cloudly = p[p==0]
p_sunny = len(sunny) / N
p_cloudly = len(cloudly) / N
print(p_sunny, p_cloudly)
```

```python
plt.plot(n, count, label = 'count')
#plt.scatter(n, p, label = 'p')
plt.axhline(p_sunny)
plt.legend()
plt.show()
```
